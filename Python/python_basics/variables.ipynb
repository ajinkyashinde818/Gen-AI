# Variables

## A variable is a such type of containers that stored the specific specified data such as integer,string, etc. is known as a variable.

## A variable is a named memory location that stores a value which can change during program execution.


```python
name="Sakshi"
age=18
salary=500.23
status=True
print("Name:", name, "\nAge:", age, "\nSalary:", salary, "\nStatus:", status)
```

## Print:- The print function/command is used to pass a data to a console to show a data 

```python
print("Name is :", name)
```

## type() :- A type() is a function that tells about a variable datatype. it means it gives a which type of variable are declared .

```python
print(type(name))
print(type(age))
print(type(salary))
print(type(status))
print(type(hi))
```
