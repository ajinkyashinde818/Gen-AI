```python
#Create a program for a small shop bill system.

#The program should:

#Take customer name
#Take price of 3 products
#Take quantity of each product
#Calculate:
#Total price of each product
#Overall subtotal
#18% GST
#Final bill
#If final bill is more than 5000:
#Give 10% discount
#Print complete formatted bill 

name=input("Enter customer name: ")

pr1_price=float(input("Enter price of product 1: "))
pr1_qty=int(input("Enter quantity of product 1: "))

pr2_price=float(input("Enter price of product 2: "))
pr2_qty=int(input("Enter quantity of product 2: "))

pr3_price=float(input("Enter price of product 3: "))
pr3_qty=int(input("Enter quantity of product 3: "))

product1_total = pr1_price * pr1_qty
product2_total = pr2_price * pr2_qty
product3_total = pr3_price * pr3_qty

subtotal = product1_total + product2_total + product3_total
gst = subtotal * 0.18
bill_amount = subtotal + gst

discount = 0
if bill_amount > 5000:
    discount = bill_amount * 0.10

final_bill = bill_amount - discount

print("------ SHOP BILL ------\n")
print("Customer Name:", name)
print()
print("Product1 Total:", product1_total)
print("Product2 Total:", product2_total)
print("Product3 Total:", product3_total)
print()
print("Subtotal:", subtotal)
print("GST (18%):", gst)
print()
print("Discount:", discount)
print()
print("Final Bill:", final_bill)

```

```python
'''
Student Result & Scholarship System

Create a Python program that:

Takes student name
Takes marks of 5 subjects
Calculates:
Total marks
Percentage
Average
Determines:
Pass or Fail
Grade
Scholarship eligibility
Conditions
Pass/Fail
If any subject marks < 35 → Fail
Otherwise → Pass
Grade System
Percentage	Grade
90+	A+
75–89	A
60–74	B
40–59	C
Below 40	D
Scholarship Rules
Percentage	Scholarship
95+	₹50,000
90–94	₹25,000
80–89	₹10,000
Below 80	No Scholarship  '''


name=input("Enter Your Name")

math=int(input("Enter Math Marks:"))
science=int(input("Enter Science Marks:"))
english=int(input("Enter English Marks:"))
computer=int(input("Enter Computer Marks:"))
history=int(input("Enter History Marks:"))

total_marks=math+science+english+computer+history

Avg=total_marks/5

percentage=total_marks/500*100

Result=""

if math<35:
    Result="Fail"
elif science<35:
    Result="Fail"

elif english<35:
    Result="Fail"

elif computer<35:
    Result="Fail"

elif history<35:
    Result="Fail"

else:
    Result="Pass"


## Grade ##
grade = ""

if percentage >= 90:
    grade="A+"
elif percentage >= 75 and percentage <= 89:
    grade="A"

elif percentage >= 60 and percentage <= 74:
    grade="B"

elif percentage >= 40 and percentage <= 59:
    grade="C"

else:
    grade="D"

#schlorship

schlor=""

if percentage >= 95:
    schlor="₹50,000"

elif percentage >= 90 and percentage <= 94:
    schlor="₹25,000"

elif percentage >= 80 and percentage <= 89:
    schlor="₹10,000"

else:
    schlor="No Schlorship"


print("----------------- Academic Insights----------------")
print()

print("Name is:",name)

print("Math:",math)
print("Science:",science)
print("English:",english)
print("Computer:",computer)
print("History:",history)

print()

print("Total Marks:",total_marks)

print("Average:",Avg)

print("Percentage:",percentage)

print("Result:",Result)

print("The Grade:",grade)

print("Schlorship:",schlor)

```
