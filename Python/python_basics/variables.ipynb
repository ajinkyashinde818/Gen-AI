# Variables

## A variable is a such type of containers that stored the specific specified data such as integer,string, etc. is known as a variable.

## A variable is a named memory location that stores a value which can change during program execution.


```python
name="Sakshi"
age=18
salary=500.23
status=True
print("Name:", name, "\nAge:", age, "\nSalary:", salary, "\nStatus:", status)
```

## Print:- The print function/command is used to pass a data to a console to show a data 

```python
print("Name is :", name)
```

## Comment:- The comment is a statement that tells about the program. means it can be easy to identified to a another peoples

## type() :- A type() is a function that tells about a variable datatype. it means it gives a which type of variable are declared .

```python
print(type(name))
print(type(age))
print(type(salary))
print(type(status))
print(type(hi))
```

```python
Age=18
print(type(Age))

converted_age=str(Age)
print(type(converted_age))
```

# Input Function

## A Input fuction in python is used to take a input from user to pass a computer and store it in a variable and print on console or output

```python
name=input("Enter your name: ")
```

```python
print("Hello :", name)
```

## Input function takes a input in string format until and unless we are not going explicitly take a input in integer or other format


```python
number=int(input("Enter a Number:"))
print(number)

name=str(input("Enter your name: "))
print("Hello :", name)
```

```python
height=float(input("Enter your height: "))
print("Your height is :", height)
```

```python
value=bool(input("Enter a value: "))
print(type(value))
print("The value is :", value)
```

```python
## Building a simple calculator

num1=int(input("Enter first number: "))
num2=int(input("Enter second number: "))
print("The sum is :", num1+num2)
print("The difference is :", num1-num2)
print("The product is :", num1*num2)
print("The quotient is :", num1/num2)
```
