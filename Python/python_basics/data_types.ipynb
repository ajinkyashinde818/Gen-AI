## Datatypes in Python define the kind of value stored in a variable, such as numbers, text, collections, or boolean values.



# Main Datatypes in Python

| Datatype | Description | Example |
|---|---|---|
| int | Integer numbers | 10 |
| float | Decimal numbers | 10.5 |
| str | Text/String | "Hello" |
| bool | True or False | True |
| complex | Complex numbers | 3+4j |
| list | Ordered mutable collection | [1,2,3] |
| tuple | Ordered immutable collection | (1,2,3) |
| set | Unordered unique values | {1,2,3} |
| dict | Key-value pairs | {"name":"Ajinkya"} |
| NoneType | No value | None |


## 1. Integer (`int`)

```python
x = 10
print(type(x))
```

## 2. Float (`float`)

```python
price = 99.99
print(type(price))
```

## 3. String (`str`)

```python
name = "Ajinkya"
print(type(name))
```

## 4. Boolean (`bool`)

```python
is_pass = True
print(type(is_pass))
```

## 5. Complex (`complex`)

```python
x = 3 + 4j
print(type(x))
```

## 6. List (`list`)

```python
marks = [90, 85, 70]
print(type(marks))
```

## 7. Tuple (`tuple`)

```python
data = (1, 2, 3)
print(type(data))
```

## 8. Set (`set`)

```python
nums = {1, 2, 3}
print(type(nums))
```

## 9. Dictionary (`dict`)

```python
student = {
    "name": "Ajinkya",
    "age": 18
}

print(type(student))
```

## 10. NoneType (`None`)

```python
x = None
print(type(x))
```


# Type Conversion

Convert one datatype into another.


```python
# Integer to String
x = str(10)
print(x, type(x))

# String to Integer
y = int("25")
print(y, type(y))

# Integer to Float
z = float(10)
print(z, type(z))
```

# Memory Concept

In Python:
- Every value is an object
- Every object has:
  - Type
  - Value
  - Memory location


```python
x = 10

print("Value:", x)
print("Type:", type(x))
print("Memory Address:", id(x))
```


# Object in Python

In Python, everything is an object.

An object is:
> A real instance of a datatype that contains data and behavior.

## Simple Meaning
- Object stores value
- Object has datatype
- Object can perform actions


## Integer Object

```python
x = 10

print(type(x))
print(id(x))
```

## String Object

```python
name = "Ajinkya"

print(type(name))
print(id(name))
```

## List Object

```python
nums = [1, 2, 3]

print(type(nums))
print(id(nums))
```


# Object Reference Concept

Conceptually:

x -----> [10]

`x` stores the reference (memory address) of the object.


```python
x = 10
y = x

print("x:", x)
print("y:", y)

print("x address:", id(x))
print("y address:", id(y))
```


# Final Summary

- Python automatically detects datatypes
- Everything in Python is an object
- Variables store references to objects
- Objects contain:
  - Type
  - Value
  - Memory Address
