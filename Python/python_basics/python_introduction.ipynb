# Python Introduction

Basic notes about Python syntax, semantics, and rules.

## What is Syntax?
Syntax means the rules for writing code correctly in Python.

Example:
```python
a = 5
print(a)
```


## What is Semantics?
Semantics means the meaning of the code.

Example:
```python
print("Hello")
```
This prints Hello on the screen.


# Rules of Python Syntax

## 1. Indentation
Python uses spaces/tabs to define blocks of code.

Example:
```python
if True:
    print("Correct Indentation")
```

Wrong Example:
```python
if True:
print("Wrong")
```


````
if True:
    print("Correct Indentation Example")
````

## 2. Case Sensitive
Python treats uppercase and lowercase differently.

Example:
```python
name = "Ajinkya"
Name = "Python"

print(name)
print(Name)
```


````
name = "Ajinkya"
Name = "Python"

print(name)
print(Name)
````

## 3. Use Backslash (\)
Backslash is used to continue code on the next line.

Example:
```python
1 + 2 + 3 + 4 + 5 + \
6 + 7
```


````
result = 1 + 2 + 3 + 4 + 5 + \
6 + 7

print(result)
````

## 4. Variables
Variables store data.

Example:
```python
a = 5
b = 6
c = 7
```


````
a = 5
b = 6
c = 7

print(a, b, c)
````

## 5. Syntax Error Example

Wrong:
```python
a=5 b=6
```

Correct:
```python
a=5
b=6
```


# Summary

- Syntax = Rules of writing code
- Semantics = Meaning of code
- Python is case sensitive
- Indentation is important
- Backslash is used for line continuation
