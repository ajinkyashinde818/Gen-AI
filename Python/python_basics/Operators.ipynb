
# Types of Operators

| Operator Type | Purpose |
|---|---|
| Arithmetic | Mathematical operations |
| Assignment | Assign values |
| Comparison | Compare values |
| Logical | Combine conditions |
| Identity | Check object identity |
| Membership | Check membership |
| Bitwise | Binary operations |


# 1. Arithmetic Operators

| Operator | Meaning        | Example   |
| -------- | -------------- | --------- |
| `+`      | Addition       | `10 + 5`  |
| `-`      | Subtraction    | `10 - 5`  |
| `*`      | Multiplication | `10 * 5`  |
| `/`      | Division       | `10 / 5`  |
| `//`     | Floor Division | `10 // 3` |
| `%`      | Modulus        | `10 % 3`  |
| `**`     | Power          | `2 ** 3`  |


```python
x = 10
y = 3

print("Addition:", x + y)
print("Subtraction:", x - y)
print("Multiplication:", x * y)
print("Division:", x / y)
print("Floor Division:", x // y)
print("Modulus:", x % y)
print("Power:", x ** y)
```

# 2. Assignment Operators



```python
x = 10

x += 5
print("x += 5:", x)

x -= 2
print("x -= 2:", x)

x *= 2
print("x *= 2:", x)

x /= 2
print("x /= 2:", x)
```

# 3. Comparison Operators

```python
x = 10
y = 5

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

# 4. Logical Operators

```python
age = 18

print(age > 15 and age < 25)
print(age > 20 or age < 25)
print(not(age > 20))
```

# 5. Identity Operators

```python
x = [1, 2]
y = x
z = [1, 2]

print(x is y)
print(x is z)
print(x is not z)
```

# 6. Membership Operators

```python
nums = [1, 2, 3, 4]

print(2 in nums)
print(10 in nums)
print(5 not in nums)
```

# 7. Bitwise Operators

```python
x = 5
y = 3

print(x & y)
print(x | y)
print(x ^ y)
print(~x)
print(x << 1)
print(x >> 1)
```

# Operator Precedence

```python
x = 10 + 5 * 2

print(x)
```
